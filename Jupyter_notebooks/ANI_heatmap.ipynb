```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/Users/lszydlowski/Desktop/jc-0055_fastani.csv')

# Create a pivot table for the heatmap
heatmap_data = df.pivot(index='query_genome', columns='reference_genome', values='ANI_value')

# Create the heatmap using the YBG color palette
plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)

# Add titles and labels
plt.title('Heatmap of ANI Values')
plt.xlabel('Reference Genome')
plt.ylabel('Query Genome')

# Display the heatmap
plt.show()
```

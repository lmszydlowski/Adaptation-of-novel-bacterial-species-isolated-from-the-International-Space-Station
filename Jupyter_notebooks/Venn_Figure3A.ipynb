```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles
import pandas as pd
import numpy as np
```

```python
# Create the Venn diagram

vd2 = venn2(subsets = (496, 376, 6126), set_labels=('A. burdickii', 'A. agilis sp. 4041'), set_colors=("yellow", "indigo"), alpha=0.4)
for text in vd2.subset_labels:  #change number size
 text.set_fontsize(20)
# Display the Venn diagram
plt.show()
plt.savefig(bbox_inches='tight', dpi=1200, fname='/Users/lszydlowski/Desktop/Venns_jc0161.png')
```
